```python
import pandas as pd
import numpy as np
import seaborn as sns
import os, time,json
```

```python
info = pd.read_csv('Data/superhero_info - superhero_info.csv')
superpower = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
info.head()
```

```python
superpower.head()
```

```python
info['Hero|Publisher'].str.split('|', expand = True)
```

```python
info[['Hero', 'Publisher']] = info['Hero|Publisher'].str.split('|', expand = True)
info.head(2)
```

```python
info = info.drop(columns ='Hero|Publisher')
info.head(2)
```

```python
info.loc[0,'Measurements']
```

```python
info['Measurements'] = info['Measurements'].str.replace("'",'"')
```

```python
info['Measurements']
```

```python
info['Measurements']= info['Measurements'].apply(json.loads)
info.loc[0, 'Measurements']
```

```python
height_weight = info['Measurements'].apply(pd.Series)
height_weight
```

```python
info = pd.concat((info, height_weight),axis = 1)
info.head(2)
```

```python
info = info.drop(columns =['Measurements'])
info.head()
```

```python
info['Height'] = info['Height'].str.replace(' cm', '')
info.head()
```

```python
info['Weight']= info['Weight'].str.replace(' kg', '')
info.head()
```

```python
info['Weight'] = pd.to_numeric(info['Weight'])
info['Height'] = pd.to_numeric(info['Height'])
```

```python
info.info()
```

```python
superpower['Powers'].head(2)
```

```python
superpower.loc[0,'Powers']
```

```python
superpower['powers_split']= superpower['Powers'].str.split(',', expand = False)
superpower['powers_split']
```

```python
exploded = superpower.explode('powers_split')
```

```python
exploded
```

```python
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    superpower[col] = superpower['Powers'].str.contains(col)
superpower.head()
```

```python
merged = pd.merge(info, superpower, left_on='Hero', right_on='hero_names')
merged.head()
```

```python
merged.groupby('Super Speed')['Height'].mean()
```

```python
sns.barplot(data= merged,x= 'Super Speed', y= 'Weight')
```

```python
merged.groupby('Publisher')['Height'].mean()
```

```python
ax = sns.barplot(data=merged,x= 'Publisher', y= 'Height')
ax.set_xticklabels(ax.get_xticklabels(), rotation= 90);
```
